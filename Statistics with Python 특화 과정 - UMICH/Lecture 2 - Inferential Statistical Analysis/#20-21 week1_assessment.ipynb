2. Write a function that inputs a list of integers and returns the minimum value

````python
# Write your function here
def get_min(lst):
    min_val = lst[0]
    for i in lst[1:]:
        if i < min_val:
            min_val = i
    return min_val
````

````python
# Test your function with input lst
lst = [-3, 0, 2, 100, -1, 2]
get_min(lst)

# Create you own input list to test with
lst2 = [-20.1, 32, 0, -1, 301, -1003]
get_min(lst2)
````
